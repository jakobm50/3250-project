**Description:** I will be integrating the data I downloaded/scraped into one dataset.

Import libraries:

```python
import pandas as pd
```

<br>

Import data:

```python
# load data into dataframes
# odds
nfl_odds = pd.read_csv("../../data/clean/clean_kaggle_data.csv", encoding = "UTF-8")

# nfl data
nfl_data = pd.read_csv("../../data/clean/clean_fixture_download.csv", encoding = "UTF-8")
```

```python
display(nfl_odds)
display(nfl_data)
```

<br>

Integrate the datasets:

```python
full_nfl_data = nfl_odds.merge(nfl_data, how = "inner", on = ["date", "location", "home_team", "away_team"])
```

```python
display(full_nfl_data)
```

<br>

Export as csv:

```python
full_nfl_data.to_csv("../../data/final/nfl_data_for_analysis.csv", encoding = "UTF-8", index = False)
```
